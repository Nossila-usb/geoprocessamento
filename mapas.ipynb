ma aula de hoje, vamos trabalar com **geoprocessamento**, que nada mais é do que am exibição de dados em mapas geograficos. Na célula de codigo abeixo, vamos importar a biblioteca de geração de mapas geográficos. Para isso vamos fazer uso da biblioteca **folium**. Paa usarmos, precisamoos instalá-la através do comando ***pip install folium***

```python
import folium
```

Na próxima célula de codigo, vamos gerar um mapa  e faze-lo abris na cidade de **Brasilia-df**

```python
mapa = folium.Map(location = [-15.7834301996, -47.889], zoom_start = 13,)
display(mapa)
```

Na proxima célula de codgo vamos criar marcações de lugares. 

```python

folium.Marker(location = [-15.7948, -47.91319], popup='SENAI SIG.\nOnde esta acontcendo o curso de Python Developer!',icon=folium.Icon(color='red')).add_to(mapa)
folium.Marker(location = [-15.7834301996, -47.899], popup='Estadio Nacional Mané Garrincha.',icon=folium.Icon(color='black')).add_to(mapa)
display(mapa)
```

Salvando o mapa em página HTML:

```python
mapa.save("index.hatml")
```
